#1- Bubblesort

```python
#1- Implementation du Bubblesort
def bubblesort(sequencia):
    n = len(sequencia)
    swapped = True
    while swapped:
        swapped = False
        for i in range(1, n):
            if sequencia[i - 1] > sequencia[i]:
                sequencia[i - 1], sequencia[i] = sequencia[i], sequencia[i - 1]
                swapped = True
        n -= 1
    return sequencia
```

#3- Insertionsort

```python
#3- Implementation du Insertionsort
def insertion_sort(lista):
    for i in range(1, len(lista)):
        current_value = lista[i]
        position = i

        while position > 0 and lista[position - 1] > current_value:
            lista[position] = lista[position - 1]
            position -= 1

        lista[position] = current_value

    return lista
```

#4- Quicksort


```python
#4- Implementation du Quicksort
def quicksort(arr):
    if len(arr) <= 1:
        return arr

    pilha = [(0, len(arr) - 1)]

    while pilha:
        inicio, fim = pilha.pop()

        pivot = arr[(inicio + fim) // 2]
        i, j = inicio, fim

        while i <= j:

            while arr[i] < pivot:
                i += 1

            while arr[j] > pivot:
                j -= 1

            if i <= j:
                arr[i], arr[j] = arr[j], arr[i]
                i += 1
                j -= 1

        if inicio < j:
            pilha.append((inicio, j))
        if i < fim:
            pilha.append((i, fim))

    return arr

```

#7- Bucketsort

```python
#7- Implementation du Bucketsort
def bucket_sort(array):
  max_value = max(array)

  buckets = [[] for _ in range(max_value + 1)]

  for item in array:
    buckets[item].append(item)

  for bucket in buckets:
    bucket.sort()

  sorted_array = []
  for bucket in buckets:
    sorted_array += bucket

  return sorted_array
```
