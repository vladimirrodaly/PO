```python
import matplotlib.pyplot as plt
import random
import math
```

**Faça um gráfico contendo a curva log2(n) e em seguida gere números aleatórios nos intervalos (0,N) onde N será (1000,2000,3000,...10000). Depois calcule o número de comparações que o computador faz para encontrar este número na sua respectiva lista e plote os gráficos conjuntamente.**

```python
# Função para gerar números aleatórios em um intervalo (0, N)
def gera_numeros_aleatorios(N):
    return random.randint(0, N)
```

```python
# Função para calcular o número de comparações para encontrar um número em uma lista
def numero_de_comparacoes(lista, numero):
    comparacoes = 0
    for elemento in lista:
        comparacoes += 1
        if elemento == numero:
            return comparacoes
    return comparacoes
```

```python
# Valores de N
N_valores = list(range(1000, 11000, 1000))

# Lista para armazenar o número de comparações
comparacoes = []
```

```python
# Gere os números aleatórios e calcule o número de comparações
for N in N_valores:
    lista = [gera_numeros_aleatorios(N) for _ in range(N)]
    numero = gera_numeros_aleatorios(N)
    comp = numero_de_comparacoes(lista, numero)
    comparacoes.append(comp)

# Crie o gráfico da curva log2(n)
x = [math.log2(N) for N in N_valores]
```

**Resultado do plots**

```python
# Plote os gráficos
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, comparacoes, marker='o', linestyle='-', color='b')
plt.xlabel('log2(N)')
plt.ylabel('Número de Comparações')
plt.title('Número de Comparações em Função de log2(N)')
```

```python
plt.subplot(1, 2, 2)
plt.plot(N_valores, comparacoes, marker='o', linestyle='-', color='r')
plt.xlabel('N')
plt.ylabel('Número de Comparações')
plt.title('Número de Comparações em Função de N')

plt.tight_layout()
plt.show()
```
