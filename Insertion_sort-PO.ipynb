```python
#decorator (medir o tempo)
def medirTempo(funcao):
  def calcula_tempo(tamanho):
    inicio = perf_counter()
    res = funcao(tamanho)
    fim = perf_counter()
    return res, fim-inicio
  return calcula_tempo
```

```python
#gerar serie aleatoria em python
@medirTempo
def geraSerieAleatoria(tam):
  s = np.arange(0, tam, 1)
  np.random.shuffle(s)
  return s
```

```python
s, t = geraSerieAleatoria(10)
print(f"Serie : {s}")
print(f"Tempo {t} segundos")
```

```python
#gerar um serie totalmente invertida
@medirTempo
def geraSerieInvertida(tam):
  return np.arange(tam-1, -1, -1)
```

```python
s, t = geraSerieInvertida(10)
print(f"Serie : {s}")
print(f"Tempo {t} segundos")
```

```python
@medirTempo
def bubble_sort(array):
  n = len(array)
  for i in range(n):
    for j in range(0, n-i-1):
      if array[j] > array[j+1]:
        array[j], array[j+1] = array[j+1], array[j]
  return array
```

```python
@medirTempo
def bubbleSort(S):
  mudanca = True
  while(mudanca):
    mudanca = False
    for i in range(len(S) - 1):
      if S[i] > S[i+1]:
        S[i], S[i+1] = S[i+1], S[i]
        mudanca = True
  return S
```

 # insertion sort

```python
import numpy as np
from time import perf_counter
from seaborn import pointplot, set_style
from pandas import DataFrame

LISTA_DE_TAMANHOS = np.array([1000, 2000, 3000,
                              4000, 5000, 8000,
                              11_000, 15_000], dtype=int)
```

```python
#ImplantaÃ§Ã£o do Insertion Sort
def insertion(arr:'np.ndarray[float]') -> None:
  n = len(arr)
  if n > 1:
    for i in range(1, n):
      atual = arr[i]
      j = i - 1
      while j >= 0 and arr[j] > atual:
        arr[j+1] = arr[j]
        j -= 1
      arr[j+1] = atual
```

**Desempenho**

```python
data = dict()

for ordem in ['crescente', 'aleatoria', 'decrescente']:
  tempos = list()
  for tamanho in LISTA_DE_TAMANHOS:
    # Gerando amostra do teste
    if ordem == 'crescente':
      amostra = np.arange(tamanho, dtype=float)
    elif ordem == 'decrescente':
      amostra = np.arange(tamanho, dtype=float)[::-1].copy()
    else:
      amostra = np.random.sample(tamanho)

    # testando performance
    start = perf_counter()
    insertion(amostra)
    stop = perf_counter()

    # salvando resultado
    tempos.append(stop - start)
  data[ordem] = tempos
data['tamanhos'] = LISTA_DE_TAMANHOS

df = DataFrame(data=data)
df.style.hide_index()
```

**Plotar o grafico & Resultado**

```python
# Remodela o dataframe para uma versÃ£o longa mais fÃ¡cil de plotar
long_df = df.melt(id_vars='tamanhos', var_name='ordenaÃ§Ã£o', value_name='tempos')

set_style('darkgrid')
pointplot(
  data=long_df,
  x='tamanhos',
  y='tempos',
  hue='ordenaÃ§Ã£o',
  ).set(
    xlabel='tamanho da amostra',
    ylabel='tempo (s)',
    title='TAMANHO DA LISTA X TEMPO DE ORDENAMENTO')

ordem = long_df.sort_values(by='tempos', ascending=False)['ordenaÃ§Ã£o'].values[0]
print(f'O pior caso Ã© quando a lista se encontra ordenada de forma {ordem}')
```
