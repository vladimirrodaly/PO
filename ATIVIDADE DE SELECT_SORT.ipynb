```python
import numpy as np
from timeit import default_timer as Timer
from timeit import timeit
import time
from random import randint
import matplotlib.pyplot as plt
import copy
```

```python
# Implementação do Selection Sort 1
```

```python
def select_sort1(array):
  for k in range(len(array)):
    menor = array[k]
    indiceDoMenor = k
    for j in range(k+1,len(array)):
      if array[j] < menor:
        indiceDoMenor = j
        menor = array[j]
    array[k],array[indiceDoMenor] = menor, array[k]
  return array
```

```python
# Implementação do Selection Sort 2
```

```python
def select_sort2(array):

  for k in range(len(array)):
    menor = array[k]
    for j in range(k+1,len(array)):
      if array[j] < array[k]:
        array[k],array[j] = array[j], array[k]

  return array
```

```python
#Verificacao de tempo de excucao
```

```python
def time_sort_function(sort_function, array):
    start_time = time.time()
    sort_function(array.copy())
    return time.time() - start_time
```

```python
#Define o tamanho da lista
```

```python
sizes = [100, 200, 400, 800, 1600, 2000, 5000, 10000]
times_select_sort1 = []
times_select_sort2 = []
```

```python
# funcao qu everificar o dados e tempos
```

```python
for n in sizes:
    random_array = [randint(0, 1000000) for _ in range(n)]

    times_select_sort1.append(time_sort_function(select_sort1, random_array))
    times_select_sort2.append(time_sort_function(select_sort2, random_array))
```

```python
# Plotar o gráfico dos dois casos
```

```python
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10, 6))
plt.plot(sizes, times_select_sort1, label='select_sort1', marker='o')
plt.plot(sizes, times_select_sort2, label='select_sort2', marker='x')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo (s)')
plt.title('Desempenho do Selection Sort')
plt.legend()
plt.show()
```
