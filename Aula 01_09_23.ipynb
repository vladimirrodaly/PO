```python
import numpy as np
from timeit import default_timer as Timer
from timeit import timeit
import time
from random import randint
import matplotlib.pyplot as plt
import copy
```

```python
#metodo 1
def select_sort(array):
  for k in range(len(array)):
    menor = array[k]
    indiceDoMenor = k
    for j in range(k+1,len(array)):
      if array[j] < menor:
        indiceDoMenor = j
        menor = array[j]
    array[k],array[indiceDoMenor] = menor, array[k]
  return array
```

```python
#metodo2
def select2_sort(array):

  for k in range(len(array)):
    menor = array[k]
    for j in range(k+1,len(array)):
      if array[j] < array[k]:
        array[k],array[j] = array[j], array[k]

  return array
```

```python
# Função para medir o tempo de execução de cada método
def time_sort_function(sort_function, array):
    start_time = time.time()
    sort_function(array.copy())
    return time.time() - start_time


```

```python
sizes = [100, 200, 400, 800, 1600, 2000, 5000, 10000]
times_select_sort = []
times_select2_sort = []

# Gerar dados e tempos
for n in sizes:
    random_array = [randint(0, 1000000) for _ in range(n)]

    times_select_sort.append(time_sort_function(select_sort, random_array))
    times_select2_sort.append(time_sort_function(select2_sort, random_array))

# Plotar o gráfico
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10, 6))
plt.plot(sizes, times_select_sort, label='select_sort', marker='o')
plt.plot(sizes, times_select2_sort, label='select2_sort', marker='x')
plt.xlabel('Tamanho do Array')
plt.ylabel('Tempo (s)')
plt.title('Comparação do Tempo de Execução de dois métodos de Selection Sort')
plt.legend()
plt.show()
```
