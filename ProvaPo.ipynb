**Buble sort**

```python
def bubblesort(sequencia):
    n = len(sequencia)
    swapped = True
    while swapped:
        swapped = False
        for i in range(1, n):
            if sequencia[i - 1] > sequencia[i]:
                sequencia[i - 1], sequencia[i] = sequencia[i], sequencia[i - 1]
                swapped = True
        n -= 1
    return sequencia
```

**Selection sort**

```python
def selection_sort(sequencia):
    n = len(sequencia)
    for i in range(n):
        for j in range(i + 1, n):
            if sequencia[i] > sequencia[j]:
                sequencia[i], sequencia[j] = sequencia[j], sequencia[i]
    return sequencia
```

**Insertion sort**

```python
def insertion_sort(lista):
    for i in range(1, len(lista)):
        current_value = lista[i]
        position = i

        while position > 0 and lista[position - 1] > current_value:
            lista[position] = lista[position - 1]
            position -= 1

        lista[position] = current_value

    return lista
```

**Quick sort**

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr

    pilha = [(0, len(arr) - 1)]

    while pilha:
        inicio, fim = pilha.pop()

        pivot = arr[(inicio + fim) // 2]
        i, j = inicio, fim

        while i <= j:

            while arr[i] < pivot:
                i += 1

            while arr[j] > pivot:
                j -= 1

            if i <= j:
                arr[i], arr[j] = arr[j], arr[i]
                i += 1
                j -= 1

        if inicio < j:
            pilha.append((inicio, j))
        if i < fim:
            pilha.append((i, fim))

    return arr
```

**Merge sort**

```python
def mergesort(lista):
    tamanho_da_lista = len(lista) - 1

    for posicao_atual in range(0, tamanho_da_lista):
        posicao_menor = posicao_atual
        menor_nome = lista[posicao_menor]

        for posicao_busca in range(posicao_atual, tamanho_da_lista):
            nome_busca = lista[posicao_busca + 1]

            if menor_nome > nome_busca:
                menor_nome = nome_busca
                posicao_menor = posicao_busca + 1

        if posicao_menor != posicao_atual:
            menor_nome = lista[posicao_menor]
            lista[posicao_menor] = lista[posicao_atual]
            lista[posicao_atual] = menor_nome

    return lista
```

**Counting sort**

```python
def counting_sort(arr):
    max_val = max(arr)
    count_arr = [0] * (max_val + 1)

    for num in arr:
        count_arr[num] += 1
    sorted_arr = []
    for i in range(len(count_arr)):
        sorted_arr.extend([i] * count_arr[i])
    return sorted_arr
```

**Bucket sort**

```python
def bucket_sort(array):
  max_value = max(array)

  buckets = [[] for _ in range(max_value + 1)]

  for item in array:
    buckets[item].append(item)

  for bucket in buckets:
    bucket.sort()

  sorted_array = []
  for bucket in buckets:
    sorted_array += bucket

  return sorted_array
```
