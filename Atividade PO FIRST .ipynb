```python
#Função para calcular pi usando simulação
```

```python
def monte_carlo(n):
  simuMC = []
  n_valores = np.arange(1, 100)
  for i in range(n):
            x = np.random.uniform(0.0, 1.0, i)
            y = np.random.uniform(0.0, 1.0, i)
            distancia = x**2 + y**2
            distancia_certa = np.sqrt(distancia)
            pontos_dentro_circulo = np.count_nonzero(distancia_certa <= 1)
  pi = (pontos_dentro_circulo*4)/len(distancia_certa)
           # simuMC.append(pi)
  return pi
```

```python
#Função para calcular pi usando a série de Leibniz
```

```python
#def calcular_pi_leibniz(n_termos):
def leibniz(tamanho):
  soma = 1
  denominador = 3
  for k in range(tamanho):
   soma -= 1.0/denominador
   denominador += 2
   soma += 1.0/denominador
   denominador +=2
  pi = 4*soma
  return pi
```

```python
#Função para calcular pi usando a série de Wijngaarden
```

```python
#def calcular_pi_wijngaarden(n_termos):
def calcular_pi_wijngaarden(ind):
  tam = 2*(ind+1)
  pi = []
  linha = []
  for i in range(tam):
    soma = 0
    denominador = 1
    j = 0
    while j <= i:
        if j%2 == 0:
            soma+=4.0/denominador
            denominador+=2
        else:
            soma-=4.0/denominador
            denominador+=2
        j+=1
    linha.append(soma)
  pi.append(linha)
  for j in range(1,tam):
    linha2 = []
    k = 0
    while k < len(linha) -1:
        linha2.append((linha[k]+linha[k+1])/2)
        k+=1
    pi.append(linha2)
    linha = linha2
  return pi[tam//2-1][tam//2-1]
```

#Biblioteca numpy para calcular a derivada numérica de qualquer função.




```python
import numpy as np
```

```python
def calcular_derivada_numerica(funcao, x, h=1e-5):
    derivada = (funcao(x + h) - funcao(x - h)) / (2 * h)
    return derivada
```

```python
#Exemplo de função para testar a função de cálculo da derivada
```

```python
def funcao_exemplo(x):
    return x**2 + 2*x + 1
```

```python
#Ponto onde queremos calcular a derivada
```

```python
ponto = 2.0
```

```python
#Calcula a derivada numérica da função de exemplo no ponto especificado
```

```python
derivada_numerica = calcular_derivada_numerica(funcao_exemplo, ponto)

print(f"A derivada numérica da função no ponto {ponto} é aproximadamente: {derivada_numerica}")
```

#Calculo da integral numérica

```python
#Calculo da integral numérica
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_integral_numerica(funcao, a, b, n_pontos=1000):
    x = np.linspace(a, b, n_pontos)
    y = funcao(x)
    dx = (b - a) / (n_pontos - 1)
    integral = (np.sum(y) - 0.5 * (y[0] + y[-1])) * dx
    return integral
```

```python
#Exemplo de função para testar o cálculo da integral numérica
```

```python
def funcao_exemplo(x):
  return x**2 + 2*x + 1
```

```python
#Intervalo de integração
```

```python
a = 0.0
b = 2.0
```

```python
#Calcula a integral numérica da função de exemplo no intervalo especificado
```

```python
integral_numerica = calcular_integral_numerica(funcao_exemplo, a, b)
```

```python
print(f"O valor da integral numérica da função no intervalo [{a}, {b}] é aproximadamente: {integral_numerica}")
```
