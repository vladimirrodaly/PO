```python
def desenha_grafico(ax, dados1, dados2, dicionario_parametros):
  saida = ax.plot(dados1,dados2,**dicionario_parametros)
  return saida
```

```python
dados1, dados2, dados3, dados4 = np.random.randn(4,100)
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(7,3))
desenha_grafico(ax1,dados1,dados2, {'marker' : 'x'})
desenha_grafico(ax2,dados3,dados4, {'marker' : 'o'})
```
